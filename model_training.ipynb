```python
# Sıralı olan verileri modelin daha iyi öğrenmesi için rastgele karıştırız
```

```python
from sklearn.utils import shuffle
```

```python
df_dur_shuffled = shuffle(df_dur, random_state=15)
df_başla_shuffled = shuffle(df_başla, random_state=15)
df_geri_shuffed = shuffle(df_geri, random_state=15)
df_ileri_shuffled = shuffle(df_ileri, random_state=15)
```

```python
# Tüm veriler df de toplanır ve tekrar karıştırılır
```

```python
df = pd.concat([df_ileri_shuffled,df_başla_shuffled,df_dur_shuffled,df_geri_shuffed], ignore_index=True)
df["hareket_tipi"].unique()
```

```python
df.info()
```

```python
df = shuffle(df,random_state=15)
```

```python
df.head()
```

```python
df.tail()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["hareket_tipi_encoded"] = encoder.fit_transform(df["hareket_tipi"])
df.drop("hareket_tipi", axis=1, inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["hareket_tipi_encoded"].unique()
```

```python
# model eğitimi
```

```python
from sklearn.model_selection import train_test_split
X = df.drop("hareket_tipi_encoded",axis=1)
y = df["hareket_tipi_encoded"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=15, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
score = accuracy_score(y_pred,y_test)
print("score: ",score)
print(classification_report(y_pred,y_test))
print("matris\n",confusion_matrix(y_pred,y_test))
```
